```python
# Initial the map

import folium
from folium.features import DivIcon

site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Iterate over launch_sites_df to add Circle and Marker for each launch site
for index, row in launch_sites_df.iterrows():
    # Create coordinate list
    coordinate = [row['Lat'], row['Long']]
    # print(coordinate, row['Launch Site'])  # For debugging

    # Add Circle with popup
    circle = folium.Circle(
        location=coordinate,
        radius=1000,
        color='#000000',
        fill=True
    ).add_child(folium.Popup(row['Launch Site']))
    circle.add_to(site_map)  # Add Circle to map

    # Add Marker with custom DivIcon
    marker = folium.Marker(
        location=coordinate,
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % row['Launch Site']
        )
    )
    marker.add_to(site_map)  # Add Marker to map

# Display or save the map
site_map  # Display in Jupyter Notebook
```

```python
import folium
from folium.plugins import MarkerCluster

# Pre-existing function
def get_marker_color(launch_class):
    if launch_class == 1:
        return 'green'
    else:
        return 'red'

# Add marker_color column based on success/failure
spacex_df['marker_color'] = spacex_df['class'].apply(get_marker_color)

# Initialize the folium map
site_map = folium.Map(location=[28.563197, -80.576820], zoom_start=5)

# Initialize and add MarkerCluster to the map
marker_cluster = MarkerCluster().add_to(site_map)

# Add Circle and Label Marker for each launch site
for index, row in launch_sites_df.iterrows():
    coordinate = [row['Lat'], row['Long']]

    # Circle to show launch site radius
    folium.Circle(
        location=coordinate,
        radius=1000,
        color='#000000',
        fill=True
    ).add_child(folium.Popup(row['Launch Site'])).add_to(site_map)

    # Text label marker
    folium.Marker(
        location=coordinate,
        icon=folium.features.DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12px; color:#d35400;"><b>%s</b></div>' % row['Launch Site']
        ),
        popup=row['Launch Site']
    ).add_to(site_map)

# Add Markers for each launch record (success/failure) with color
for index, row in spacex_df.iterrows():
    coordinate = [row['Lat'], row['Long']]

    folium.Marker(
        location=coordinate,
        popup=f"{row['Launch Site']} - {'Success' if row['class'] == 1 else 'Failure'}",
        icon=folium.Icon(color=row['marker_color'], icon_color='white', icon='rocket', prefix='fa')
    ).add_to(marker_cluster)

# Show the map
site_map
```

Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
launch_site_lat = 28.562302
launch_site_lon = -80.577356

coastline_lat = 28.55422
coastline_lon = -80.56652

distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

distance_marker = folium.Marker(
    [coastline_lat,coastline_lon],
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
        )
    )
```
